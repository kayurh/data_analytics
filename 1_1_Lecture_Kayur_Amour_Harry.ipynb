### Creating a DataFrame from Lists and Dictionaries

Imagine we are building a small database of a company’s employees.

How can we convert this into a DataFrame?

```python
# Import the pandas package (commonly used with the alias 'pd')
import pandas as pd

# Create a dictionary where each key maps to a list
# - 'Name' column: names of three people
# - 'Age' column: ages
# - 'City' column: places of residence
data = {'name': ['Anna', 'Béla', 'Csaba'],
        'age': [23, 45, 32],
        'city': ['Budapest', 'Szeged', 'Debrecen']}
```

```python
display(data)
```

```python
# A pandas DataFrame is used to store data in a tabular (table-like) format.
# Here, the dictionary keys become the column names, and the list elements become the row values.
df = pd.DataFrame(data)
```

```python
# Print the DataFrame.
# (In a Jupyter notebook, 'display(df)' shows the result in a nicer table format.)
display(df)
```

Creating a DataFrame from Lists

```python
# Store the data in three separate lists
names = ['Anna', 'Béla', 'Csaba']          # list of names
age = [23, 45, 32]                       # list of ages
city = ['Budapest', 'Szeged', 'Debrecen']  # list of cities
```

```python
print(names)
print(age)
print(city)
```

```python
# The zip() function "packs" the elements of the lists together:
# [('Anna', 23, 'Budapest'),
#  ('Béla', 45, 'Szeged'),
#  ('Csaba', 32, 'Debrecen')]
# That is, each row becomes a tuple (name, age, city).
# list() converts the zip object into an actual list.
list(zip(names, age, city))
```

```python
# Then we load the data into a pandas DataFrame,
# and specify the column names using the 'columns' parameter.
df = pd.DataFrame(list(zip(names, age, city)),
                  columns=['name', 'age', 'city'])
```

```python
# In a Jupyter notebook, display() shows the data in a nice table format.
display(df)
```

Creating a DataFrame from a List of Lists

```python
# Create a list where each element is another list.
# Each inner list represents one data row: [name, age, city].
adatok = [
    ['Anna', 23, 'Budapest'],
    ['Béla', 45, 'Szeged'],
    ['Csaba', 32, 'Debrecen']
]

# When creating the DataFrame:
# - 'adatok' is interpreted as rows
# - column names are provided via the 'columns' parameter.
df = pd.DataFrame(adatok, columns=['name', 'age', 'city'])

# Print the DataFrame.
# (In Jupyter, display(df) is nicer.)
print(df)
```

Creating a DataFrame from a NumPy array

If we receive the data as a NumPy array (e.g., from machine learning models or mathematical operations),
we can easily convert it using the pandas.DataFrame() constructor.

```python
import numpy as np # Import NumPy (for arrays and numerical operations)

# Create a NumPy array.
# Each row contains one person's data: [name, age, city].
numpy_array = np.array([
    ['Anna', 23, 'Budapest'],
    ['Béla', 45, 'Szeged'],
    ['Csaba', 32, 'Debrecen']
])

# Pass the NumPy array to the pandas DataFrame constructor.
# The 'columns' parameter defines the column names.
df = pd.DataFrame(numpy_array, columns=['name', 'age', 'city'])


# Print the DataFrame.
# (In a notebook, it’s better to use display(df).)
print(df)
```

Advantages:

If the source data comes as a NumPy array (e.g., ML datasets), this method is convenient.
It may be faster for large datasets.

### Creating / Loading a DataFrame from a CSV file

In real life, data is often stored in CSV files.
The pandas read_csv() function is one of the best tools for loading CSV data.

```python
# Read a CSV (Comma-Separated Values) file directly from the internet.
# pandas.read_csv() automatically downloads and loads the table as a DataFrame.
# The data can also come from any local file path.
df = pd.read_csv('https://gist.github.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv')

# head() shows the first 5 rows by default.
# This is useful for quickly checking what the table contains.
display(df.tail())  # display the first 5 rows
```

More details: https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html

### Loading a CSV with different parameters

In practice, CSV files can come in many different formats:
- they may use a different separator (e.g., ";" or tab instead of ",")
- they might not have a header, or the header may not be on the first row
- they may contain special characters, accents, or missing values marked in different ways

pandas.read_csv() provides many parameters to handle these cases.
Some commonly used settings:

- sep → field separator character
- header / names → header row index or custom column names
- index_col → which column should be used as the index
- usecols → load only specific columns
- dtype → enforce data types
- na_values → specify which values should be treated as missing
- skiprows, nrows → skip rows / read only a limited number of rows
- encoding → file encoding (important for special characters)

In the example below, we demonstrate how these parameters can be used.


```python
# CSV loading with useful parameters (example)
df = pd.read_csv(
    'https://gist.github.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv',  # file path or URL
    sep=",",               # column separator (e.g., ";" or "\t" are also possible)
    header=0,              # which row to use as the header (0 = first row)
    names=None,            # if there is no header, we can provide column names, e.g. ['A','B','C']
    index_col=None,        # which column should be used as the index (e.g., 'ID')
    usecols=None,          # load only specific columns, e.g. ['sepal.length','variety']
    dtype=None,            # enforce data types, e.g. {'sepal.length': float, 'variety': 'category'}
    na_values=["?", "NA"], # values to treat as missing data
    skiprows=0,            # number of rows to skip at the start of the file
    nrows=None,            # number of rows to read (e.g., for testing: nrows=100)
    encoding="utf-8"       # file encoding (sometimes needed e.g. "latin1" or "cp1250" for special characters)
)

# Quick check
print(df.head())

```

### The Titanic dataset

The Titanic dataset is one of the most well-known open datasets.
It is often used for practicing data processing, statistical analysis, and machine learning.
The data contain information about passengers of the RMS Titanic, which sank on April 15, 1912,
and the goal is to analyze who survived the disaster.

Dataset characteristics
- Source: OpenML (based originally on Kaggle and other statistical publications)
- Number of rows: 1309 passengers
- Number of columns: 14 features

Dataset columns
- pclass = passenger class (1 = first, 2 = second, 3 = third class)
- sex = passenger gender (male/female)
- age = passenger age (some values are missing)
- sibsp = number of siblings/spouses aboard
- parch = number of parents/children aboard
- fare = ticket price (in pounds)
- embarked = port of embarkation (C=Cherbourg, Q=Queenstown, S=Southampton)
- home.dest = passenger home or destination
- survived = target variable (0 = did not survive, 1 = survived)

Interesting facts about Titanic survival:
- First-class passengers had a much higher survival rate than third-class passengers.
- Women had significantly higher chances of survival than men.
- Due to the "women and children first" rule, younger passengers were more likely to survive.

```python
# Load Titanic data using scikit-learn

from sklearn.datasets import fetch_openml
import pandas as pd

# Load the dataset from OpenML
titanic = fetch_openml(name="titanic", version=1, as_frame=True)

# Convert data to a Pandas DataFrame
df = titanic.data
df["survived"] = titanic.target  # Add the target variable (y)
```

### Exercise 3
- Create a new column where you build categories from the AGE column:
  "child (<18), adult (18–65), senior (>65)"
- Count how many women and men traveled on the ship by age group
- Show (in %) how many women and men traveled on the ship by age group
- Count by travel class / age group how many women and men traveled on the ship

```python
import pandas as pd
from sklearn.datasets import fetch_openml

titanic = fetch_openml(name='titanic', version=1, as_frame=True)
df = titanic.frame

df['age_group'] = pd.cut(df['age'], bins=[0, 18, 65, 120], labels=['child', 'adult', 'senior'])
print(df.groupby(['age_group', 'sex']).size())
print(df.groupby(['age_group', 'sex']).size().groupby(level=0).apply(lambda x: x / x.sum() * 100))
print(df.groupby(['pclass', 'age_group', 'sex']).size())
```

### Exercise 4
The "survived" column shows whether someone survived the disaster.
- Show (in %) how many women and men survived
- Show by age group (in %) how many women and men survived
- Show by travel class (in %) how many passengers survived
- Show by travel class and gender (in %) how many passengers survived

```python
import pandas as pd
from sklearn.datasets import fetch_openml

titanic = fetch_openml(name='titanic', version=1, as_frame=True)
df = titanic.frame

df['survived'] = df['survived'].astype(int)
df['age'] = pd.to_numeric(df['age'], errors='coerce')

df['age_group'] = pd.cut(df['age'], bins=[0, 18, 65, 120], labels=['child', 'adult', 'senior'])

print(df.groupby('sex')['survived'].mean() * 100)
print(df.groupby(['age_group', 'sex'])['survived'].mean() * 100)
print(df.groupby('pclass')['survived'].mean() * 100)
print(df.groupby(['pclass', 'sex'])['survived'].mean() * 100)
```

### Exercise 5
The "survived" column shows whether someone survived the disaster.
- Show the average age of women and men
- Compare the age distribution of survivors vs non-survivors
- Compare the age distribution of survivors vs non-survivors by gender
- Compare the age distribution of survivors vs non-survivors by age group

```python
import pandas as pd
from sklearn.datasets import fetch_openml

titanic = fetch_openml(name='titanic', version=1, as_frame=True)
df = titanic.frame


df['survived'] = df['survived'].astype(int)
df['age'] = pd.to_numeric(df['age'], errors='coerce')

df['age_group'] = pd.cut(df['age'], bins=[0, 18, 65, 120], labels=['child', 'adult', 'senior'])

print(df.groupby('sex')['age'].mean())
print(df.groupby('survived')['age'].describe())
print(df.groupby(['survived', 'sex'])['age'].describe())
print(df.groupby(['survived', 'age_group'])['age'].describe())
```
